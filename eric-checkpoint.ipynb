```python
#séparer la variable cible COEUR 
X = df.iloc[:,:11]
X
```

```python
#stocker la variable COEUR dans un objet appelé Y 
Y = data.iloc[:,11:]
Y
```

```python
# repatition de nos données en données d'entrainement train et de test

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
np.array(Y_test)
```

```python
#instanciation du modèle
model = linear_model.LogisticRegression(random_state=2, solver = 'liblinear', multi_class = 'auto')
model.fit(X_train, Y_train)
```

```python
prediction = model.predict_proba(X_test)
print('probabilte du modèle:', prediction)
```

```python
#prediction
prediction = model.predict(X_test)
print('prédiction du modèle:', prediction)
```

```python
#calcule de précision
confusion_matrix(Y_test, prediction)
```

```python
accuracy_score(Y_test,prediction)
```

```python
recall_score(Y_test,prediction)
```

```python
precisio = precision_score(Y_test, prediction)
print('la précision du modèle', precisio )
```
