```python
import os
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.linear_model import LogisticRegression
```

```python
#1- importation du jeu de Données
data=pd.read_excel(r'C:\Users\BADJO M\Downloads\exo-regression\Coeur.xlsx')
```

```python
# 2-Copiez l’objet data dans un nouvel objet df 
df = data.copy()
df
```

```python
#3.Vérifiez s’il y a des doublons dans l’objet df 
df.duplicated()
```

```python
#Supprimer les doublons
df.drop_duplicates(keep=False)
```

```python
#4.Vérifier s’il y a des constantes 
df.isna()
```

```python
#5.Vérifier s’il y a des données manquantes dans df 
df.isnull()
```

```python
#supprimer les données manquants
df.dropna()
```

```python
#6.normalisation
normalized_df=(df-df.mean())/df.std()
normalized_df
```

```python
#7.Faite un recodage de toutes les variables qualitatives.
def recoder(df):
    for i in df.select_dtypes('object').columns:
        df[i]=df[i].astype('category').cat.codes
    return(df)
recoder(df)
```

```python
#séparer la variable cible COEUR 
X = df.iloc[:,:11]
X
```

```python
#stocker la variable COEUR dans un objet appelé Y 
Y = data.iloc[:,11:]
Y
```

```python
# repatition de nos données en données d'entrainement train et de test

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 0)
```

```python
#instanciation du modèle
modele_regLog = linear_model.LogisticRegression(random_state=0, solver = 'liblinear', multi_class = 'auto')
```

```python
#training
modele_regLog.fit(X_train, Y_train)
```

```python
#prediction
prediction = modele_regLog.predict(X_test)
print('prédiction du modèle:', prediction)
```

```python
#calcule de précision
precision = modele_regLog.score(X_test, prediction)
print('la précision du modèle', precision * 100)
```
